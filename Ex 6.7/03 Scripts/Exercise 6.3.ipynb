### Plotting a choropleth

```python
# Create a dataframe
data_to_plot = df[['Location','Rating']]
data_to_plot.head()
```

```python
import folium

# Create map
map = folium.Map(location=[20, 0], zoom_start=2)

# Add choropleth
folium.Choropleth(
    geo_data=country_geo, 
    data=data_to_plot,
    columns=['Location', 'Rating'],  # numeric column for coloring
    key_on='feature.properties.name',
    fill_color='YlOrBr',
    fill_opacity=0.6,
    line_opacity=0.1,
    legend_name="Chocolate Rating"
).add_to(map)

# Add layer control
folium.LayerControl().add_to(map)

# Display map
map
```

Highest-rated regions: Switzerland, Canada, and the U.K. show the highest average ratings, indicating strong consumer satisfaction and possibly higher-quality chocolate production.
